## Visualization & Explanatory Analysis

### Pertanyaan 1:

#### Bagaimana kondisi penyewaan sepeda pada tahun 2011?

```python
data_df = bike_hour_df.copy()
```

```python
filter_df = data_df[data_df["yr"] == 0]
```

```python
filter_df.set_index('dteday', inplace=True)
```

```python
rentals_df = filter_df.resample(rule='ME').agg({
    "cnt": "sum"
})

rentals_df.reset_index(inplace=True)

plt.figure(figsize=(10, 5)) 
plt.plot(rentals_df["dteday"], rentals_df["cnt"], marker='o', linewidth=2, color="#72BCD4") 
plt.title("Jumlah Penyewaan Sepeda pada Tahun 2011", loc="center", fontsize=20) 
plt.xticks(rentals_df["dteday"], rentals_df["dteday"].dt.strftime('%B %Y'), rotation=30,fontsize=10) 
plt.yticks(fontsize=10) 
plt.ylabel("Jumlah Penyewa")
plt.grid(True)
plt.show()
```

```python
# Menentukan bulan dengan total penyewaan tertinggi dan terendah
bulan_tertinggi = rentals_df["cnt"].idxmax()
bulan_terendah = rentals_df["cnt"].idxmin()

highest_month = rentals_df.iloc[bulan_tertinggi]
lowest_month = rentals_df.iloc[bulan_terendah]

# print hasil
print(f"Bulan dengan penyewaan tertinggi: {highest_month['dteday'].strftime('%B %Y')} dengan total penyewaan {highest_month['cnt']}")
print(f"Bulan dengan penyewaan terendah: {lowest_month['dteday'].strftime('%B %Y')} dengan total penyewaan {lowest_month['cnt']}")
```

#### Pada tahun 2011 penyewaan sepeda terendah terdapat pada bulan Januari dengan total 38189 dan penyewaan tertinggi pada bulan Juni dengan total 143512, setelah itu penyewaan sepeda mengalami penurunan hingga akhir Desember 2021.

### Pertanyaan 2:

#### Pada musim apa penyewaan sepeda terbanyak?

```python
# Menghitung total penyewaan per season
total_rentals = data_df.groupby('season')['cnt'].sum().reset_index()

# Menentukan season dengan penyewaan terbanyak
max_musim = total_rentals.loc[total_rentals['cnt'].idxmax()]

# Menampilkan hasil
print(f"Season dengan penyewaan terbanyak: {max_musim['season']} dengan total penyewaan {max_musim['cnt']}")

# Membuat daftar warna untuk setiap bar
colors = ['skyblue'] * len(total_rentals)  # Warna default untuk semua bar
max_index = total_rentals['cnt'].idxmax()  # Indeks bar dengan penyewaan tertinggi
colors[max_index] = 'blue'  # Ubah warna bar tertinggi

# Membuat barplot
plt.figure(figsize=(10, 5))
plt.title('Total Penyewaan Sepeda Setiap Musim')
bars = plt.bar(total_rentals['season'], total_rentals['cnt'], color=colors)
plt.xlabel('Musim')
plt.ylabel('Total Penyewaan')

# Menambahkan nilai total penyewaan di atas setiap bar
for bar in bars:
    yval = bar.get_height()  # Mendapatkan tinggi bar
    plt.text(bar.get_x() + bar.get_width()/2, yval, int(yval), ha='center', va='bottom')  # Menambahkan teks

    # Mengatur label sumbu X secara manual
plt.xticks(ticks=total_rentals['season'], labels=total_rentals['season'])

# Mengatur batas sumbu Y
plt.ylim(0, total_rentals['cnt'].max() * 1.1)  # Menambahkan sedikit ruang di atas bar

# Mengatur format angka pada sumbu Y
def millions(x, pos):
    return '%1.0f' % (x)

formatter = FuncFormatter(millions)
plt.gca().yaxis.set_major_formatter(formatter)

plt.show()
```

#### Penyewaan sepeda terbanyak terjadi pada musim musim gugur dengan total penyewaan sebanyak 1061129

### Pertanyaan 3:

#### Bagaimana perbandingan penyewaan sepeda di hari kerja dengan akhir pekan?

```python
# Menghitung total penyewaan untuk hari kerja dan akhir pekan
total_rentals = data_df.groupby('workingday')['cnt'].sum().reset_index()

# Mengubah nilai workingday menjadi label
total_rentals['day_type'] = total_rentals['workingday'].map({0: 'Akhir Pekan', 1: 'Hari Kerja'})

# Menyiapkan data untuk pie chart
sizes = total_rentals['cnt']
labels = total_rentals['day_type']

# Modifikasi label untuk menampilkan jumlah total penyewaan
labels_jumlah = [f"{label}: {count}" for label, count in zip(labels, sizes)]

explode = (0, 0.1)  # Memisahkan bagian akhir pekan

# Membuat pie chart
plt.figure(figsize=(10, 8))
plt.pie(sizes, labels=labels_jumlah, autopct='%1.1f%%', startangle=90, explode=explode, colors=['grey', 'yellow'])
plt.legend(labels, loc='upper right', frameon=False)
plt.title('Perbandingan Penyewaan Sepeda di Hari Kerja dengan Akhir Pekan')
plt.show()
```

#### Pada pie chart diatas ditunjukkan bahwa penyewaan sepeda pada hari kerja 69.6 % dibanding dengan akhir pekan 30.4%

- Conclution pertanyaan 2 : Penyewaan sepeda terbanyak terjadi pada musim musim gugur dengan total penyewaan sebanyak 1061129 dan penyewaan sepeda terendah terjadi pada musim semi.
